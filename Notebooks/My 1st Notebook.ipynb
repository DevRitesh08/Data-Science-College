```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,np.pi*2,360)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(sin,'r-.',label='sin')
plt.plot(cos,'g-.',label='cos')
plt.plot([0,360],[0,0],'k-.')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
x = np.arange(1,101)
y = np.random.randint(1,50,100)
plt.scatter(x,y,c='r',ec='g')
plt.plot(x,y,'g-.')
plt.show()
```
